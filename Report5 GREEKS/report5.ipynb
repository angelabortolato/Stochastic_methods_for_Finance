```python
import numpy as np
import matplotlib.pyplot as plt

# =========================
# 1. LOAD DATA
# =========================

# Surfaces (option prices)
Z_01 = np.loadtxt("Cprice01.txt")   # vol = 0.1
Z_02 = np.loadtxt("Cprice02.txt")   # vol = 0.2
Z_03 = np.loadtxt("Cprice03.txt")   # vol = 0.3

# Axes
S = np.loadtxt("S.txt")     # underlying prices (rows)
T = np.loadtxt("T.txt")     # maturities (columns)

# =========================
# 2. CREATE GRID
# =========================
vmin = min(Z_01.min(), Z_02.min(), Z_03.min())
vmax = max(Z_01.max(), Z_02.max(), Z_03.max())

T_grid, S_grid = np.meshgrid(T, S)
z_offset = vmin  # bottom plane

# =========================
# 3. CREATE FIGURE
# =========================

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# =========================
# 4. PLOT SURFACES
# =========================

# surfaces (simple)
ax.plot_surface(S_grid, T_grid, Z_01, color='blue', alpha=0.35)
ax.plot_surface(S_grid, T_grid, Z_02, color='green', alpha=0.35)
ax.plot_surface(S_grid, T_grid, Z_03, color='red', alpha=0.35)


# 5. LABELS & TITLE
# =========================
ax.invert_yaxis()
ax.invert_xaxis()
ax.set_xlabel("Underlying Price (S)")
ax.set_ylabel("Time to Maturity (T)")

ax.set_title("Call Price Surfaces for Different Volatilities")

# =========================
# 6. VIEW ANGLE (TUNE THIS)
# =========================

ax.view_init(elev=20, azim=45)
# =========================
# 7. LEGEND (manual)
# =========================

from matplotlib.patches import Patch

legend_elements = [
    Patch(facecolor='blue', alpha=0.5, label='Vol = 0.1'),
    Patch(facecolor='green', alpha=0.5, label='Vol = 0.2'),
    Patch(facecolor='red', alpha=0.5, label='Vol = 0.3')
]

ax.legend(handles=legend_elements, loc='upper left')

# =========================
# 8. SHOW PLOT
# =========================

plt.tight_layout()
plt.show()
```
